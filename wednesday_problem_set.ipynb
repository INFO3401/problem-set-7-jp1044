### Problem 4

```python
df = data[data['Question'] == "Percent of adults aged 18 years and older who have obesity"]
df = df[df['Data_Value'].notnull()==True]['Data_Value']
sb.distplot(df)
```

This distribution follows a normal distribution. It tells us that the average percentage of adults that are obese is 30 percent. It also tells us that they are generally an equal amount of people above and below this average.

### Problem 5

```python
df_q = data[data['Question'] == "Percent of adults aged 18 years and older who have obesity"]
sb.factorplot('YearStart', 'Data_Value', data=df_q)
sb.jointplot('YearStart', 'Data_Value', df_q)
```

This distribution indicates that obesity rates have followed the same normal distribution between 2011 and 2016. It also indicates that the average obesity rate has increased between 2011 and 2016.
